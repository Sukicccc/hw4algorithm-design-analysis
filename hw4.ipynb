- å‡è®¾å­˜åœ¨å¦ä¸€æ£µç”Ÿæˆæ ‘ \(T'\)ï¼Œå…¶æœ€å¤§è¾¹æƒæ›´å°ï¼Œé‚£ä¹ˆ \(T'\) ä¸åŒ…å« \(T_{MST}\) çš„æœ€å¤§æƒé‡è¾¹ï¼Œè¿åäº†MSTæ€§è´¨ï¼ˆå³æœ€å°è½»è¾¹å¿…é¡»è¢«åŒ…å«ï¼‰ã€‚

#### ç®—æ³•æ€è·¯ï¼š
- æœ¬è´¨ï¼šåˆ¤æ–­å›¾ä¸­æ˜¯å¦å­˜åœ¨ä¸€ä¸ª**ç”Ÿæˆæ ‘**ï¼Œæ‰€æœ‰è¾¹æƒ \(\leq b\)ã€‚
- ç­‰ä»·äºï¼šåœ¨åˆ å»æ‰€æœ‰æƒé‡ > b çš„è¾¹åï¼Œåˆ¤æ–­å›¾æ˜¯å¦ä»è¿é€šã€‚
- è¿é€šæ€§æ£€æŸ¥å¯ä»¥ç”¨ **DFS** æˆ– **BFS**ï¼Œæ—¶é—´å¤æ‚åº¦ \(O(|V| + |E|)\)ã€‚

```python
# add your code here
def exists_bottleneck_spanning_tree(G, b):
    from collections import defaultdict, deque
    
    V = len({u for u, v, w in G} | {v for u, v, w in G})  # è·å–èŠ‚ç‚¹æ•°
    adj = defaultdict(list)
    
    # åªä¿ç•™æƒé‡ä¸è¶…è¿‡ b çš„è¾¹
    for u, v, w in G:
        if w <= b:
            adj[u].append(v)
            adj[v].append(u)
    
    # BFS è¿é€šæ€§æ£€æµ‹
    visited = set()
    def bfs(start):
        queue = deque([start])
        visited.add(start)
        while queue:
            node = queue.popleft()
            for nei in adj[node]:
                if nei not in visited:
                    visited.add(nei)
                    queue.append(nei)

    bfs(next(iter(adj)))  # ä»ä»»æ„ä¸€ä¸ªå­˜åœ¨çš„èŠ‚ç‚¹å¼€å§‹
    return len(visited) == V  # åˆ¤æ–­æ˜¯å¦æ‰€æœ‰ç‚¹éƒ½è¿é€š

G = [
    (0, 1, 4),
    (0, 2, 3),
    (1, 2, 2),
    (1, 3, 5),
    (2, 3, 1)
]

print(exists_bottleneck_spanning_tree(G, 3))  # True
print(exists_bottleneck_spanning_tree(G, 2))  # False

# algorithm of the liear time complexity ï¼šğ‘‚(âˆ£Vâˆ£+âˆ£Eâˆ£)
```

3.ç®—æ³•æ€è·¯ï¼š
ç±»ä¼¼äº MSTï¼Œä½†å…³æ³¨æœ€å¤§è¾¹æƒã€‚
å¯ä»¥ç”¨äºŒåˆ†æ³•ï¼šåœ¨æ‰€æœ‰è¾¹æƒä¸ŠäºŒåˆ†æŸ¥æ‰¾æœ€å°ç“¶é¢ˆå€¼ã€‚
å¯¹äºæ¯ä¸ªå€™é€‰è¾¹æƒbï¼Œç”¨ç¬¬2é—®çš„ç®—æ³•åˆ¤æ–­æ˜¯å¦å­˜åœ¨ç”Ÿæˆæ ‘ã€‚
æœ€ç»ˆç”¨ Kruskalç®—æ³• æ”¶é›†ç“¶é¢ˆç”Ÿæˆæ ‘çš„è¾¹ã€‚

```python
def bottleneck_spanning_tree(G, V):
    # è·å–æ‰€æœ‰è¾¹æƒ
    weights = sorted(set([w for _, _, w in G]))
    
    # äºŒåˆ†æŸ¥æ‰¾æœ€å°ç“¶é¢ˆå€¼
    def can_form_spanning_tree(b):
        from collections import defaultdict, deque
        adj = defaultdict(list)
        for u, v, w in G:
            if w <= b:
                adj[u].append(v)
                adj[v].append(u)
        visited = set()
        def dfs(node):
            visited.add(node)
            for nei in adj[node]:
                if nei not in visited:
                    dfs(nei)
        dfs(0)
        return len(visited) == V
    
    left, right = 0, len(weights) - 1
    res = weights[right]
    while left <= right:
        mid = (left + right) // 2
        if can_form_spanning_tree(weights[mid]):
            res = weights[mid]
            right = mid - 1
        else:
            left = mid + 1
    
    # ç”¨Kruskalæ”¶é›†ç“¶é¢ˆç”Ÿæˆæ ‘
    parent = [i for i in range(V)]
    def find(x):
        while parent[x] != x:
            parent[x] = parent[parent[x]]
            x = parent[x]
        return x
    def union(x, y):
        parent[find(x)] = find(y)
    
    T = []
    edges = sorted([e for e in G if e[2] <= res], key=lambda x: x[2])
    for u, v, w in edges:
        if find(u) != find(v):
            union(u, v)
            T.append((u, v, w))
            if len(T) == V - 1:
                break
    return T
# algorithm of the liear time complexity ï¼šğ‘‚((V+E)logE)
```

### ã€æ—¶é—´å¤æ‚åº¦åˆ†æã€‘

1. **ä¸¤æ¬¡ Dijkstra æœ€çŸ­è·¯å¾„**
   - ä» `s` å‡ºå‘ä¸€æ¬¡
   - ä» `t` å‡ºå‘ä¸€æ¬¡ï¼ˆåœ¨åå‘å›¾ä¸Šï¼‰
   
   æ¯æ¬¡å¤æ‚åº¦ä¸ºï¼š$$O(m \log n)$$
   æ‰€ä»¥ä¸¤æ¬¡æ˜¯ï¼š$$O(2 \cdot m \log n) = O(m \log n)$$

2. **éå†å€™é€‰æ–°è¾¹åˆ—è¡¨ $E'$**
   - å‡è®¾å€™é€‰è¾¹æ•°ä¸º $m'$
   - æ¯æ¡è¾¹è®¡ç®—ä¸€æ¬¡è·ç¦»æ”¹è¿›é‡ï¼Œæ€»å¤æ‚åº¦ï¼š$$O(m')$$

### **æ€»æ—¶é—´å¤æ‚åº¦**
$$O(m \log n + m')$$

å…¶ä¸­ï¼š
- $n$ = èŠ‚ç‚¹æ•°
- $m$ = åŸå›¾çš„è¾¹æ•°
- $m'$ = å€™é€‰æ–°è¾¹çš„æ•°é‡

```python
import heapq
from collections import defaultdict

def dijkstra(n, edges, start):
    graph = defaultdict(list)
    for u, v, w in edges:
        graph[u].append((v, w))

    dist = [float('inf')] * n
    dist[start] = 0
    heap = [(0, start)]

    while heap:
        d, u = heapq.heappop(heap)
        if d > dist[u]:
            continue
        for v, w in graph[u]:
            if dist[v] > dist[u] + w:
                dist[v] = dist[u] + w
                heapq.heappush(heap, (dist[v], v))

    return dist

def best_new_road(n, E, E_prime, s, t):
    dist_s = dijkstra(n, E, s)
    reversed_E = [(v, u, w) for u, v, w in E]
    dist_t = dijkstra(n, reversed_E, t)

    best_reduction = 0
    best_edge = None

    original_dist = dist_s[t]

    for u, v, w in E_prime:
        if dist_s[u] != float('inf') and dist_t[v] != float('inf'):
            new_dist = dist_s[u] + w + dist_t[v]
            reduction = original_dist - new_dist
            if reduction > best_reduction:
                best_reduction = reduction
                best_edge = (u, v)

    return best_edge, best_reduction

# æµ‹è¯•
n = 5
E = [
    (0, 1, 2),
    (1, 2, 2),
    (2, 3, 2),
    (3, 4, 2)
]

E_prime = [
    (0, 2, 1),
    (1, 4, 1),
    (0, 3, 10)
]

s = 0
t = 4

best_edge, best_reduction = best_new_road(n, E, E_prime, s, t)

print(f"åº”é€‰æ‹©æ–°è¾¹ {best_edge}ï¼Œå¯å‡å°‘è·ç¦» {best_reduction}")

```

###  æœ€ç»ˆç®—æ³•æ€»å¤æ‚åº¦

```python
import networkx as nx

def escape_problem(n, start_points, grid_points):
    """
    è§£å†³é€ƒç¦»é—®é¢˜ï¼šæ˜¯å¦å­˜åœ¨ m æ¡ä¸ç›¸äº¤è·¯å¾„ä»ç»™å®šèµ·ç‚¹é€ƒç¦»åˆ°è¾¹ç•Œã€‚
    
    å‚æ•°:
    - n: ç½‘æ ¼çš„å¤§å°ï¼Œn x n
    - start_points: èµ·ç‚¹åˆ—è¡¨ï¼Œæ¯ä¸ªèµ·ç‚¹æ˜¯ä¸€ä¸ª (x, y) å…ƒç»„
    - grid_points: ç½‘æ ¼ä¸­çš„æ‰€æœ‰ç‚¹çš„é›†åˆ [(i, j)]
    
    è¿”å›:
    - æ˜¯å¦èƒ½å¤Ÿæ‰¾åˆ° m æ¡ä¸ç›¸äº¤è·¯å¾„
    """
    
    # æ„å»ºä¸€ä¸ªæµç½‘ç»œå›¾
    G = nx.DiGraph()
    
    # åˆ›å»ºæºç‚¹å’Œæ±‡ç‚¹
    source = 'source'
    sink = 'sink'
    
    # åˆ›å»ºä¸€ä¸ªæ–°çš„å›¾ï¼Œå°†æ¯ä¸ªèŠ‚ç‚¹åˆ†ä¸ºå…¥ç‚¹å’Œå‡ºç‚¹ (v_in, v_out)
    for (i, j) in grid_points:
        node_in = f'{i}_{j}_in'
        node_out = f'{i}_{j}_out'
        
        # ä¸ºæ¯ä¸ªèŠ‚ç‚¹å¢åŠ é¡¶ç‚¹å®¹é‡
        G.add_edge(node_in, node_out, capacity=1)
        
        # ä¸ºå››ä¸ªæ–¹å‘çš„é‚»å±…åŠ ä¸Šè¾¹
        for di, dj in [(-1, 0), (1, 0), (0, -1), (0, 1)]:
            ni, nj = i + di, j + dj
            if 0 <= ni < n and 0 <= nj < n:  # ç¡®ä¿åœ¨ç½‘æ ¼å†…
                neighbor_in = f'{ni}_{nj}_in'
                G.add_edge(node_out, neighbor_in, capacity=1)
    
    # è¿æ¥æºç‚¹ä¸èµ·ç‚¹
    for (x, y) in start_points:
        node_in = f'{x}_{y}_in'
        G.add_edge(source, node_in, capacity=1)
    
    # è¿æ¥æ±‡ç‚¹ä¸è¾¹ç•Œç‚¹
    for i in range(n):
        for j in [0, n-1]:
            node_out = f'{i}_{j}_out'
            G.add_edge(node_out, sink, capacity=1)
        for j in range(n):
            for i in [0, n-1]:
                node_out = f'{i}_{j}_out'
                G.add_edge(node_out, sink, capacity=1)
    
    # æ±‚æœ€å¤§æµ
    flow_value, flow_dict = nx.maximum_flow(G, source, sink)
    
    # å¦‚æœæµå€¼å¤§äºç­‰äºèµ·ç‚¹æ•°mï¼Œè¡¨ç¤ºå¯ä»¥é€ƒç¦»
    return flow_value >= len(start_points)

# ç¤ºä¾‹è°ƒç”¨ï¼š
n = 6  # ç½‘æ ¼å¤§å°
start_points = [(1, 1), (2, 2), (3, 3)]  # èµ·ç‚¹åˆ—è¡¨
grid_points = [(i, j) for i in range(n) for j in range(n)]  # æ‰€æœ‰ç‚¹åˆ—è¡¨

result = escape_problem(n, start_points, grid_points)
if result:
    print("é€ƒç¦»æˆåŠŸï¼Œæ‰¾åˆ°è¶³å¤Ÿçš„è·¯å¾„ï¼")
else:
    print("æ— æ³•é€ƒç¦»ï¼Œè·¯å¾„ä¸å¤Ÿï¼")
```
